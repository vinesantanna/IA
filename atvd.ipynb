```python
%matplotlib inline
import matplotlib.pyplot as plt
dfNona.hist(bins=30, figsize=(10,15))
plt.show()
```

```python
dfNona['ocean_proximity'].unique()
y= dfNona.replace({'NEAR BAY':0, '<1H OCEAN':1, 'INLAND':2, 'NEAR OCEAN':3,'ISLAND':4})
y['ocean_proximity'].unique()
```

```python
cols1 = df.columns[0:10]
y[cols1] = y[cols1].apply(pd.to_numeric, errors='coerce')
y['ocean_proximity'].unique()
```

```python
yNona = y.dropna(axis="rows")
yNona
```

```python
yNona.info()
```

```python
yN = yNona.drop(columns=['latitude', 'longitude'])
yN.info()
```

```python
yNout = yN
cleanCols = yNout.columns[:7]
for i in cleanCols:
    yNout = yNout.loc[yNout[i] < yNout[i].quantile(0.95)]
yNout
```

```python
features = list(yNout.columns[0:8])
features

x = yNout[features]

x = yNout.drop(columns=['median_house_value'])

y = yNout['median_house_value']

%matplotlib inline
import matplotlib.pyplot as plt
yNout.hist(bins=30, figsize=(10,15))
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=6)
```

```python
regr = GradientBoostingRegressor()

model = regr.fit(x_train, y_train)

#y_pred = regr.predict(x_test)

score = model.score(x_test, y_test)
score=round(score*100,0)
print score, '%'
```
